```python
h_dt = pd.read_csv('HighPrice.csv')
l_dt = pd.read_csv('LowPrice.csv')
o_dt = pd.read_csv('OpenPrice.csv')
v_dt = pd.read_csv('TradeValue.csv')
```

```python
def choose_stock(stock_code):
    df = pd.DataFrame()
    df['open'] = o_dt[stock_code]
    df['close'] = p_dt[stock_code]
    df['high'] = h_dt[stock_code]
    df['low'] = l_dt[stock_code]
    df['trade value'] = v_dt[stock_code]
    df['turnover'] = to_dt[stock_code]
    df['return'] = 0
    for i in range(len(df)-1):
        df.iloc[i+1,6] = df.iloc[i+1,1]/df.iloc[i,1]-1
    return df
```

```python
def clean_stock(df):
    df = df.dropna()                 #drop the NAN value 
    df = df[df['turnover'] > 0]      #drop the 0 turnover value
    df.index = range(len(df))        #Reindex
    return df
```

```python
code1 = '600519.SH'
code2 = '600019.SH'
code3 = '601398.SH'
code4 = '300431.SZ'
MaoTai = clean_stock(choose_stock(code1))
BaoGang = clean_stock(choose_stock(code2))
ICBC = clean_stock(choose_stock(code3))
BaoFeng = clean_stock(choose_stock(code4))
```

```python
MaoTai.to_csv('600519.csv')
BaoGang.to_csv('600019.csv')
ICBC.to_csv('601398.csv')
BaoFeng.to_csv('300431.csv')
```
